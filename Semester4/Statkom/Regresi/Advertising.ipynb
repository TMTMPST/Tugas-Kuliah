```python
# Import Library
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_excel("Advertisings.xlsx")
df.head()
```

```python
# Menentukan X dan Y
X = df[['TV', 'radio', 'newspaper']]
y = df['sales']
```

```python
# Buat dan Latih Model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
# Evaluasi Model
intercept = model.intercept_
coefficients = model.coef_
r2 = r2_score(y, y_pred)
mse = mean_squared_error(y, y_pred)

print("Intercept:", intercept)
print("Koefisien:", coefficients)
print("R²:", r2)
print("MSE:", mse)
```

```python
# Visualisasi: Prediksi vs Aktual
plt.figure(figsize=(8,6))
sns.scatterplot(x=y, y=y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], '--r')
plt.xlabel("Penjualan Aktual")
plt.ylabel("Penjualan Prediksi")
plt.title("Prediksi vs Aktual - Regresi Linear")
plt.grid(True)
plt.show()

# Line Fit Plots for each advertising channel
plt.figure(figsize=(18, 5))

# TV Line Fit Plot
plt.subplot(1, 3, 1)
sns.regplot(x=X['TV'], y=y, line_kws={'color': 'red'})
plt.title('TV Advertising vs Sales (Line Fit Plot)')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')

# Radio Line Fit Plot
plt.subplot(1, 3, 2)
sns.regplot(x=X['radio'], y=y, line_kws={'color': 'red'})
plt.title('Radio Advertising vs Sales (Line Fit Plot)')
plt.xlabel('Radio Advertising Budget')
plt.ylabel('Sales')

# Newspaper Line Fit Plot
plt.subplot(1, 3, 3)
sns.regplot(x=X['newspaper'], y=y, line_kws={'color': 'red'})
plt.title('Newspaper Advertising vs Sales (Line Fit Plot)')
plt.xlabel('Newspaper Advertising Budget')
plt.ylabel('Sales')

plt.tight_layout()
plt.show()
```
